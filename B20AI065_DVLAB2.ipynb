##APARTMENT PRICE VS GROUND LIVING AREA.

```python
df=pd.read_csv("https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/2_TwoNum.csv")
df['SalePrice']=df['SalePrice'].apply(lambda x: x/1000)
df
```

###distribution

```python
plt.title("Ground Living Area Distribution",fontsize=12)
plt.hist(df.GrLivArea,bins=np.arange(0,5000,100),color="#69b3a2",rwidth=0.9,zorder=3)
plt.grid(zorder=0)
plt.xlabel("area")
plt.ylabel("count")
plt.xticks([0,1000,2000,3000,4000,5000])
plt.yticks([0,50,100])
plt.show()
```

```python
plt.title("Sale Price Distribution",fontsize=12)
plt.hist(df.SalePrice,bins=np.arange(0,700,20),color="#69b3a2",rwidth=0.9,zorder=3)
plt.grid(zorder=0)
plt.xlabel("Sale price (k$)")
plt.ylabel("count")
plt.xticks([0,200,400,600])
plt.yticks([0,50,100,150])
plt.show()
```

###scatterplot

```python
df.plot.scatter(y = 'SalePrice', x = 'GrLivArea', color="#69b3a2")
plt.title("Ground living area partially explains sale price of apartments", size=12)
plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
plt.yticks([0,200,400,600])
plt.xticks([0,1000,2000,3000,4000,5000])
plt.show()
```

###regression

```python
df.plot.scatter(y = 'SalePrice', x = 'GrLivArea', color="#69b3a2")
sns.regplot(data=df, x='GrLivArea', y='SalePrice', color="black", ci=None, scatter=False)
plt.title("Ground living area partially explains sale price of apartments", size=12)
plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
plt.yticks([0,200,400,600])
plt.xticks([0,1000,2000,3000,4000,5000])
plt.show()
```

###overplotting

```python
df.plot.scatter(y = 'SalePrice', x = 'GrLivArea', color="#69b3a2",s=25)
plt.title("Dot Size", size=12)
plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
plt.yticks([0,200,400,600])
plt.xticks([0,1000,2000,3000,4000,5000])
plt.show()
```

```python
sns.kdeplot(x=df.GrLivArea, y=df.SalePrice, color="#69b3a2")
plt.title("Density 2d: contour", size=12)
plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
plt.ylim(0,500)
plt.yticks([0,100,200,300])
plt.xticks([0,1000,2000])
plt.show()
```

```python
sns.kdeplot(x=df.GrLivArea, y=df.SalePrice, color="#69b3a2", cmap="Blues_r",shade=True)
plt.title("Density 2d: area", size=12)
plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
plt.xlim(0,3000)
plt.ylim(0,500)
plt.yticks([0,100,200,300])
plt.xticks([0,1000,2000])
plt.show()
```

```python
plt.hist2d(x=df.GrLivArea, y=df.SalePrice, bins=(50, 50), cmap="BuPu")
plt.title("2d histogram", size=12)
plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
plt.yticks([0,200,400,600])
plt.xticks([0,1000,2000,3000,4000,5000])
plt.show()
```

```python
plt.hexbin(x=df.GrLivArea, y=df.SalePrice, cmap="BuPu")
plt.title("HexBin", size=12)
plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
plt.yticks([0,200,400,600])
plt.xticks([0,1000,2000,3000,4000,5000])
plt.show()
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/7_OneCatOneNum.csv")
df=df.dropna()
```

###barplot


```python
plt.barh(y=df.Country, width=df.Value);
```

###circular bar plot

```python
plt.figure(figsize=(20,10))
ax = plt.subplot(111, polar=True)
plt.axis('off')
upperLimit = 10000
lowerLimit = 3000
labelPadding = 4

max = df['Value'].max()
slope = (max - lowerLimit) / max
heights = slope * df.Value 
width = 2*np.pi / len(df.index)

indexes = list(range(1, len(df.index)+1))
angles = [element * width for element in indexes]
angles

bars = ax.bar(
    x=angles, 
    height=heights, 
    width=width, 
    bottom=lowerLimit,
    linewidth=2, 
    edgecolor="white",
    color="#61a4b2",
)

for bar, angle, height, label in zip(bars,angles, heights, df["Country"]):

    rotation = np.rad2deg(angle)

    alignment = ""
    if angle >= np.pi/2 and angle < 3*np.pi/2:
        alignment = "right"
        rotation = rotation + 180
    else: 
        alignment = "left"
    ax.text(
        x=angle, 
        y=lowerLimit + bar.get_height() + labelPadding, 
        s=label, 
        ha=alignment, 
        va='center', 
        rotation=rotation, 
        rotation_mode="anchor") 
```

###circle packing

```python
pip install circlify
```

```python
import circlify
```

```python
circles = circlify.circlify(
    df['Value'].tolist(), 
    show_enclosure=False, 
    target_enclosure=circlify.Circle(x=0, y=0, r=1))
```

```python
print(circles)
```

```python
fig, ax = plt.subplots(figsize=(10,10))

# Remove axes
ax.axis('off')

# Find axis boundaries
lim = max(
    max(
        abs(circle.x) + circle.r,
        abs(circle.y) + circle.r,
    )
    for circle in circles
)
plt.xlim(-lim, lim)
plt.ylim(-lim, lim)

# print circles
for circle in circles:
    x, y, r = circle
    ax.add_patch(plt.Circle((x, y), r, alpha=0.2, linewidth=2, fill=False))

for circle in circles:
    if circle.level != 2:
      continue
    x, y, r = circle
    ax.add_patch( plt.Circle((x, y), r, alpha=0.5, linewidth=2, color="lightblue"))

# Print circle and labels for the highest level:
for circle in circles:
    if circle.level != 3:
      continue
    x, y, r = circle
    label = circle.ex["Country"]
    ax.add_patch( plt.Circle((x, y), r, alpha=0.5, linewidth=2, color="#69b3a2"))
    plt.annotate(label, (x,y ), ha='center', color="white")

# Print labels for the continents
for circle in circles:
    if circle.level != 2:
      continue
    x, y, r = circle
    label = circle.ex["Country"]
    plt.annotate(label, (x,y ) ,va='center', ha='center', bbox=dict(facecolor='white', edgecolor='black', boxstyle='round', pad=.5))
```
