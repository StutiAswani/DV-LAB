##APARTMENT PRICE VS GROUND LIVING AREA.

```python
df=pd.read_csv("https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/2_TwoNum.csv")
df['SalePrice']=df['SalePrice'].apply(lambda x: x/1000)
df
```

###distribution

```python
plt.title("Ground Living Area Distribution",fontsize=12)
plt.hist(df.GrLivArea,bins=np.arange(0,5000,100),color="#69b3a2",rwidth=0.9,zorder=3)
plt.grid(zorder=0)
plt.xlabel("area")
plt.ylabel("count")
plt.xticks([0,1000,2000,3000,4000,5000])
plt.yticks([0,50,100])
plt.show()
```

```python
plt.title("Sale Price Distribution",fontsize=12)
plt.hist(df.SalePrice,bins=np.arange(0,700,20),color="#69b3a2",rwidth=0.9,zorder=3)
plt.grid(zorder=0)
plt.xlabel("Sale price (k$)")
plt.ylabel("count")
plt.xticks([0,200,400,600])
plt.yticks([0,50,100,150])
plt.show()
```

###scatterplot

```python
df.plot.scatter(y = 'SalePrice', x = 'GrLivArea', color="#69b3a2")
plt.title("Ground living area partially explains sale price of apartments", size=12)
plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
plt.yticks([0,200,400,600])
plt.xticks([0,1000,2000,3000,4000,5000])
plt.show()
```

###regression

```python
df.plot.scatter(y = 'SalePrice', x = 'GrLivArea', color="#69b3a2")
sns.regplot(data=df, x='GrLivArea', y='SalePrice', color="black", ci=None, scatter=False)
plt.title("Ground living area partially explains sale price of apartments", size=12)
plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
plt.yticks([0,200,400,600])
plt.xticks([0,1000,2000,3000,4000,5000])
plt.show()
```

###overplotting

```python
df.plot.scatter(y = 'SalePrice', x = 'GrLivArea', color="#69b3a2",s=25)
plt.title("Dot Size", size=12)
plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
plt.yticks([0,200,400,600])
plt.xticks([0,1000,2000,3000,4000,5000])
plt.show()
```

```python
sns.kdeplot(x=df.GrLivArea, y=df.SalePrice, color="#69b3a2")
plt.title("Density 2d: contour", size=12)
plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
plt.xlim(0,3000)
plt.ylim(0,500)
plt.yticks([0,100,200,300])
plt.xticks([0,1000,2000])
plt.show()
```

```python
sns.kdeplot(x=df.GrLivArea, y=df.SalePrice, color="#69b3a2", cmap="Blues_r",shade=True)
plt.title("Density 2d: area", size=12)
plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
plt.xlim(0,3000)
plt.ylim(0,500)
plt.yticks([0,100,200,300])
plt.xticks([0,1000,2000])
plt.show()
```

```python
plt.hist2d(x=df.GrLivArea, y=df.SalePrice, bins=(50, 50), cmap="BuPu")
plt.title("2d histogram", size=12)
plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
plt.yticks([0,200,400,600])
plt.xticks([0,1000,2000,3000,4000,5000])
plt.show()
```

```python
plt.hexbin(x=df.GrLivArea, y=df.SalePrice, cmap="BuPu")
plt.title("HexBin", size=12)
plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
plt.yticks([0,200,400,600])
plt.xticks([0,1000,2000,3000,4000,5000])
plt.show()
```

##EVOLUTION OF THE BITCOIN PRICE

```python
df=pd.read_csv("https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/3_TwoNumOrdered.csv",delimiter=' ')
df
```

##line plot

```python
df["date"] = pd.to_datetime(df["date"])
plt.plot(df['date'],df.value,color="#69b3a2")
plt.title("Evolution of Bitcoin price", size=12)
plt.xlabel('date')
plt.ylabel('Bitcoin price ($)')
plt.show()
```

###area chart

```python
df["date"] = pd.to_datetime(df["date"])
plt.plot(df['date'],df.value,color="#69b3a2")
plt.fill_between(df['date'],df.value,color="#69b3a2",alpha=0.5)
plt.title("Evolution of Bitcoin price", size=12)
plt.xlabel('date')
plt.ylabel('Bitcoin price ($)')
plt.show()
```

###connected scatter

```python
import datetime
```

```python
df["date"] = pd.to_datetime(df["date"])
plt.plot(df['date'],df.value,color="#69b3a2",linestyle='-', marker='o')
plt.fill_between(df['date'],df.value,color="#69b3a2",alpha=0.5)
plt.xlim([datetime.date(2018, 4, 14), datetime.date(2018, 4, 24)])
plt.ylim([0, 10000])
plt.title("Evolution of Bitcoin price", size=12)
plt.yticks([0,2500,5000,7500])
plt.xticks([datetime.date(2018, 4, 15),datetime.date(2018, 4, 17),datetime.date(2018, 4, 19),datetime.date(2018, 4, 21),datetime.date(2018, 4, 23)])
plt.xlabel('date')
plt.ylabel('Bitcoin price ($)')
plt.show()
```

###cutting

```python
df["date"] = pd.to_datetime(df["date"])
plt.plot(df['date'],df.value,color="#69b3a2",linestyle='-', marker='o')
plt.fill_between(df['date'],df.value,color="#69b3a2",alpha=0.5)
plt.xlim([datetime.date(2018, 4, 14), datetime.date(2018, 4, 24)])
plt.ylim([8000, 9100])
plt.title("Evolution of Bitcoin price", size=12)
plt.yticks([8000,8250,8500,8750,9000])
plt.xticks([datetime.date(2018, 4, 15),datetime.date(2018, 4, 17),datetime.date(2018, 4, 19),datetime.date(2018, 4, 21),datetime.date(2018, 4, 23)])
plt.xlabel('date')
plt.ylabel('Bitcoin price ($)')
plt.show()
```

###comparing to a limit

```python
df["date"] = pd.to_datetime(df["date"])
plt.plot(df['date'],df.value,color="#69b3a2")
plt.fill_between(df['date'],df.value,where=(df.value <= 7500), interpolate=True,color="#69b3a2",alpha=0.5)
plt.xlim([datetime.date(2017, 6, 29), datetime.date(2018, 4, 12)])
plt.title("Evolution of Bitcoin price", size=12)
plt.xlabel('date')
plt.ylabel('Bitcoin price ($)')
plt.show()
```

```python
df
```

```python
names=["Ashley", "Amanda", "Jessica","Patricia", "Linda", "Deborah",   "Dorothy", "Betty", "Helen"]
```

```python
df1 = pd.DataFrame()
```

```python
df
```

```python
for i in range(0,len(df)):
  if(df.name[i] in names and df.sex[i]!='M'):
    df1=df1.append(df.loc[i],ignore_index=True)
    # print(i)
    # print(df.loc[i])
```

```python
df1
```

```python
df1.columns
```

```python
df1=df1.drop(columns=['sex', 'prop'])
```

```python
df1
```

###line plot

```python
sns.lineplot(data = df1, x = "year", y = "n", hue = "name")
plt.title("Poplarity of American names in the previous 30 years", size=12)
```

###area plot

```python
g = sns.FacetGrid(df1, col='name', hue='name', col_wrap=4, )
g = g.map(plt.plot, 'year', 'n')
g = g.map(plt.fill_between, 'year', 'n', alpha=0.2).set_titles("{col_name} name")
g = g.set_titles("{col_name}")
plt.subplots_adjust(top=0.92)
g = g.fig.suptitle('Poplarity of American names in the previous 30 years')
plt.show()
```

###stacked area chart

```python
df2=df1.pivot(index='year', columns='name', values='n')
df2
```

```python
df2.plot.area()
```

```python
data_perc = df2.divide(df2.sum(axis=1), axis=0)
```

```python
data_perc.plot.area()
```

###steamplot

```python
# fig, ax = plt.subplots(figsize=(10,5))
plt.xlim=([1880,2020])

plt.stackplot(df2.index, df2.T, baseline='wiggle', labels=df2.columns)
plt.show()
```
